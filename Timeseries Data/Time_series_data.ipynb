**Clustering -**

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

# Re-loading the data
df = pd.read_csv('/content/Stocks timeseries data.csv')

# Selecting the features for clustering
features = ['Open', 'High', 'Low', 'Close', 'Volume']
X = df[features]

# Scaling the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Applying K-means clustering
kmeans = KMeans(n_clusters=3, random_state=42)
df['Cluster'] = kmeans.fit_predict(X_scaled)

# Returning the count of data points in each cluster
cluster_counts = df['Cluster'].value_counts()
cluster_counts
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

# Re-loading the data and applying clustering
df = pd.read_csv('/content/Stocks timeseries data.csv')
features = ['Open', 'High', 'Low', 'Close', 'Volume']
X = df[features]
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
kmeans = KMeans(n_clusters=3, random_state=42)
df['Cluster'] = kmeans.fit_predict(X_scaled)
df['Date'] = pd.to_datetime(df['Date'])

# Time Series Plot
plt.figure(figsize=(15, 7))
for cluster in df['Cluster'].unique():
    subset = df[df['Cluster'] == cluster]
    plt.plot(subset['Date'], subset['Close'], label=f'Cluster {cluster}')

plt.title('Time Series of Stock Closing Price by Cluster')
plt.xlabel('Date')
plt.ylabel('Closing Price')
plt.legend()
plt.grid(True)
plt.show()
```
